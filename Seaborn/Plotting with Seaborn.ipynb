```python
import pandas as pd
apple = pd.read_csv('../Datasets/AppleStore.csv')

import seaborn as sns
```

```python
apple.columns
```

```python
sns.countplot(apple['user_rating'])
```

```python
sns.kdeplot(apple.query('price < 10').price)
```

```python
apple[apple['price']<10]['price'].value_counts().sort_index().plot.line()
```

**Distplot**: The seaborn equivalent to a pandas histogram is the distplot. 

```python
sns.distplot(apple['user_rating'], bins=10, kde=False)
```

#### Scatter plot and hexplot

To plot two variables against one another in `seaborn`, we use `jointplot`.

```python
sns.jointplot(x='size_bytes', y='user_rating', data=apple[apple['price']<10])
```

```python
sns.jointplot(x='size_bytes', y='user_rating', data=apple[apple['price']<10], kind='hex', gridsize=20)
```
