## Faceting with Seaborn

Faceting is the act of breaking data variables up across multiple subplots, and combining those subplots into a single figure. So instead of one bar chart, we might have, say, four, arranged together in a grid.

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("always")

import matplotlib.pyplot as plt
%matplotlib inline

poke = pd.read_csv('../Datasets/pokemon_lite.csv')

poke.head(3)
```

```python
# FacetGrid

g = sns.FacetGrid(poke, row="Legendary")
g.map(sns.kdeplot, "Attack")
```

```python
g = sns.FacetGrid(poke, col="Legendary", row="Generation")
g.map(sns.kdeplot, "Attack")
```

```python
# Pairplot

sns.pairplot(poke[['HP', 'Attack', 'Defense']])
```

`pairplot` is very useful for faceting variables (as opposed to variable values). You pass it a `pandas` DataFrame in the right shape, and it returns you a gridded result of variable values.

**Note**Why is faceting useful? Faceting is the easiest way to visualize multivariate data.
