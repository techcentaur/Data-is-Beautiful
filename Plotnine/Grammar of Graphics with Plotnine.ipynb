## Plotnine

Plotnine, is a data visualisation library with implements the grammar of graphics. It works by expressing every operation into a function. The **Data** element is a call to `ggplot` (by the name of library in R: `ggplot2` -- prolly an inspiration), which populates the data in the graph. The **Aesthetics** are controlled by the `aes` functions, which populates our visual variables. Finally, **Layers** are functions that add or modify the plot itself.

In a `plotnine` plot these are concatenated together with a plus (`+`) operator.

```python
import pandas as pd
import warnings

poke = pd.read_csv('../Datasets/pokemon_lite.csv')

warnings.filterwarnings("always")

poke.head(3)
```

```python
from plotnine import *
(
ggplot(poke, aes('Attack', 'Defense')) 
    + geom_point()
)

# Ignore the warnings
```

```python
(
ggplot(poke, aes('Attack', 'Defense', color='Legendary'))
    + geom_point()
    + ggtitle("Pokemon Attack and Defense Legendary Status")
)
```

```python
# We can use histograms and faceting as simply as others

(
ggplot(poke, aes('Attack'))
    + geom_histogram(bins=20)
) + facet_wrap('~Generation')
```

`Plotnine` is an amazing library to use; it's usage of functions are obvious, and smooth which makes it over the edge when compared to others. It's grammar of graphics has an ingenious approach. Really thoughtful.
